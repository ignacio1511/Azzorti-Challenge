# **Pasos Previos:**

```python
#Cargar librerías
import pandas as pd
import seaborn as sns
#Cargamos el dataset taxis
df = sns.load_dataset('taxis')
#Cambiamos el nombre de las columnas 
df.rename(columns={'pickup': 'InfoRecojo', 'dropoff': 'InfoLlegada', 
                   'passengers':'nroPasajeros', 'distance': 'distancia', 
                   'fare': 'tarifa', 'tip': 'propina', 'tolls': 'peajes',
                   'total':'total', 'color': 'colorTaxi', 'payment': 'metodoPago',
                   'pickup_zone': 'AreaRecojo', 'dropoff_zone': 'AreaLlegada',
                   'pickup_borough': 'DistritoRecojo', 'dropoff_borough': 'DistritoLlegada'}, inplace=True)
df.head(5)
```

## **PARTE A**

Pregunta: Cambiar los valores de colores 

```python
set(df['colorTaxi'])
```

```python
#Cambiar valores colorTaxi 
df["colorTaxi"].replace({"green": "verde", "yellow": "amarillo"}, inplace=True)
df.head(5)
```

Pregunta: Cambiar los métodos de Pago

```python
set(df['metodoPago'])
```

```python
#Cambiar valores metodoPago 
df["metodoPago"].replace({"cash": "contado", "credit card": "credito"}, inplace=True)
df.head(5)
```

```python
#Se puede cambiar el tamaño de las figuras 
sns.set(rc={"figure.figsize":(10, 8)}) #ancho=8, alto=4
```

```python
set(df['DistritoRecojo'])
```

Pregunta: La distribución de viajes de taxis en Manhantan

```python
dfManhatan = df[df["DistritoRecojo"] == 'Manhattan']
```

```python
sns.countplot(data=dfManhatan, x = 'nroPasajeros', hue= 'colorTaxi')
```

```python
sns.histplot(data=dfManhatan, x= 'total')
```

```python
sns.boxplot(data=dfManhatan, x='metodoPago', y='total')
```

```python
sns.scatterplot(data=dfManhatan, x='distancia', y= 'tarifa', hue='colorTaxi')
```
